```python
# Import données POC Anticor

import pandas as pd

data_marchespublics=pd.read_csv("marches_publics_normalized_data.csv", sep=";",index_col=0)

data_marchespublics.head()
```

```python
# Import données externes 
# Données codes CPV + correspondance
# Source :  https://www.achatpublic.com/actualites/la-salle-des-marches-la-recherche-par-procedure-et-codes-cpv
# Données supplémentaires ajoutées dans la colonne Division issues de la source : https://www.marche-public.fr/Marches-publics/Nomenclature/Marches-publics-Nomenclature-CPV.htm
# Fichier cpv_2008_fr enregistré en local

data_codes_cpv = pd.read_excel("cpv_2008_fr.xls", engine="xlrd")  # xlrd pour les anciens fichiers .xls
data_codes_cpv.head()
```

```python
# Ajout colonne code_simplifié pour merge les datasets

data_codes_cpv['CODE_SIMPLE'] = data_codes_cpv['CODE'].str.split('-').str[0]

data_codes_cpv.head()
```

```python
# Corrections doublons / erreurs dataset marches_publics

# Correction colonne lieu exécution type code
data_marchespublics = data_marchespublics.replace({'Code pays':'CODE PAYS','Code région':'CODE REGION','Code arrondissement':'CODE ARRONDISSEMENT','Code postal':'CODE POSTAL',
                 'Code commune' : 'CODE COMMUNE','Code canton':'CODE CANTON','Code département':'CODE DEPARTEMENT'})

# Correction colonne forme prix
data_marchespublics = data_marchespublics.replace({'Ferme, actualisable':'Ferme et actualisable'})

# Correction colonne nature
data_marchespublics = data_marchespublics.replace({'Accord-cadre':'ACCORD-CADRE','Marché':'MARCHE','Marché subséquent':'MARCHE SUBSEQUENT','Marché de partenariat':'MARCHE DE PARTENARIAT'})
```

```python
#Vérification des données

print('NB codes cpv dataset ANTICOR : ',data_marchespublics['codecpv'].nunique())
print('NB codes cpv fichier correspondances codes : ',data_codes_cpv['CODE'].nunique())
print('NB valeurs uniques Acheteurs : ',data_marchespublics['acheteur_nom'].nunique(), 'vs. NB valeurs uniques id Acheteur : ',data_marchespublics['acheteur_id'].nunique())
print('Valeurs uniques code lieu exécution : ',data_marchespublics['lieuexecution_typecode'].unique())
print('NB valeurs uniques uid : ',data_marchespublics['uid'].nunique())
print('Valeurs uniques forme prix : ',data_marchespublics['formeprix'].unique())
print('NB Valeurs uniques objet : ',data_marchespublics['objet'].nunique())
print('Valeurs uniques nature : ',data_marchespublics['nature'].unique())
print('Valeurs uniques type : ',data_marchespublics['type'].unique())
```

```python
#Vérification erreurs dates

from datetime import datetime

def detect_errors(dates):
    errors = {}
    allowed_formats = ["%Y-%m-%d"]

    for date in dates.dropna().unique():
        if isinstance(date, str):
            valid = False
            for fmt in allowed_formats:
                try:
                    parsed_date = datetime.strptime(date, fmt)
                    # Vérification de l'année (exclure des valeurs aberrantes comme 0202)
                    if 2000 <= parsed_date.year <= datetime.now().year:
                        valid = True
                        break
                except ValueError:
                    continue
            
            if not valid:
                errors[date] = errors.get(date, 0) + 1  # Stocke l'erreur et son occurrence

    return errors

errors_detected_notif = detect_errors(data_marchespublics['datenotification'])
print('Erreurs dates notification : ', errors_detected_notif)

errors_detected_pub = detect_errors(data_marchespublics['datepublicationdonnees'])
print('Erreurs dates publication : ', errors_detected_pub)
```

```python
# Modification format date incohérente
data_marchespublics = data_marchespublics.replace('September, 16 2021 00:00:00','2021-09-16')

# Suppression lignes erreurs dates
dates_incorrectes_notif = detect_errors(data_marchespublics['datenotification']).keys()
dates_incorrectes_pub = detect_errors(data_marchespublics['datepublicationdonnees']).keys()

data_marchespublics = data_marchespublics[~data_marchespublics['datenotification'].isin(dates_incorrectes_notif)]
data_marchespublics = data_marchespublics[~data_marchespublics['datepublicationdonnees'].isin(dates_incorrectes_pub)]

errors_detected_notif = detect_errors(data_marchespublics['datenotification'])
print('Erreurs dates notification : ', errors_detected_notif)

errors_detected_pub = detect_errors(data_marchespublics['datepublicationdonnees'])
print('Erreurs dates publication : ', errors_detected_pub)
```

```python
# Correction formats données

data_marchespublics = data_marchespublics.astype({'codecpv' : 'object','acheteur_id':'object','siren':'object'})
data_marchespublics['datenotification']=pd.to_datetime(data_marchespublics['datenotification'])
data_marchespublics['datepublicationdonnees']=pd.to_datetime(data_marchespublics['datepublicationdonnees'])
```

```python
# Ajout colonne avec code cpv simplifié
data_marchespublics['CODE_SIMPLE'] = data_marchespublics['codecpv'].astype(str).str[:8]
```

Ajout colonne avec type simplifié en ville, département, région et collectivité outre mer pour simplifier l'analyse

```python
# Changement nom df
data_mp = data_marchespublics

# Ajout colonne 'type simplifié' pour partager en 3 catégories : ville, département, région
data_mp['type_simpl'] = data_mp['type']

# Correction colonne 'type simplifié' pour les départements
# Data départements : type DEP, EPT (Paris)

data_mp.loc[data_mp['type'].isin(['DEP', 'EPT']), 'type_simpl'] = 'DEP'

# Correction colonne 'type simplifié' pour les villes
# Data villes : types CA, CC, COM, CU, MET, DEP(Paris)

data_mp.loc[(data_mp['nom'] == 'Paris') | (data_mp['type'].isin(['CA', 'CC', 'COM', 'CU', 'MET'])), 'type_simpl'] = 'VIL'

# Correction colonne 'type simplifié' pour les régions
# La Réunion à intégrer dans les CTU
data_mp.loc[(data_mp['nom'] == 'La Réunion'),'type_simpl']='CTU'

data_mp['type_simpl'].unique()
```

```python
# Ajout colonne année
data_mp['datenotification']=pd.to_datetime(data_mp['datenotification'])
data_mp['année_publ'] = data_mp['datenotification'].dt.year
```

```python
# Extraction dataset corrigé
data_marchespublics.to_csv('dataset_marchespublics_clean.csv',index = False)
```

```python
# Création dataframe par type

data_collectivite = data_mp.groupby(['type_simpl','année_publ']).agg({'montant':sum,'objet':n_modalities})
data_collectivite = data_collectivite.reset_index()
print(data_collectivite)
```

```python
# Création dataframe par collectivité outre mer

data_ctu = data_mp.loc[data_mp['type_simpl']=='CTU'].groupby(['nom','année_publ']).agg({'montant':sum,'objet':n_modalities})
data_ctu = data_ctu.reset_index()
print(data_ctu)
```

```python
# Création dataframe par département

data_dep = data_mp.loc[data_mp['type_simpl']=='DEP'].groupby(['nom','année_publ']).agg({'montant':sum,'objet':n_modalities})
data_dep = data_dep.reset_index()
print(data_dep)
```

```python
# Création dataframe par région

data_reg = data_mp.loc[data_mp['type_simpl']=='REG'].groupby(['nom','année_publ']).agg({'montant':sum,'objet':n_modalities})
data_reg = data_reg.reset_index()
print(data_reg)
```

```python
# Création data par ville

data_vil = data_mp.loc[data_mp['type_simpl']=='VIL'].groupby(['nom','année_publ']).agg({'montant':sum,'objet':n_modalities})
data_vil = data_vil.reset_index()
print(data_vil)
```

```python
# Création fichier excel avec les dataframes

with pd.ExcelWriter("MP_par_type.xlsx", engine="xlsxwriter") as writer:
    data_collectivite.to_excel(writer, sheet_name="MP_annee_type", index=False)
    data_vil.to_excel(writer, sheet_name="MP_detail_villes", index=False)
    data_dep.to_excel(writer, sheet_name="MP_detail_departement", index=False)
    data_reg.to_excel(writer, sheet_name="MP_detail_region", index=False)
    data_ctu.to_excel(writer, sheet_name="MP_detail_autres", index=False)

print("Fichier Excel généré avec succès 🎉")
```
