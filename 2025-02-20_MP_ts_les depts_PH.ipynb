```python
# Function to determine the 'dept' value
def determine_dept(code):
    code_str = str(code)
    if len(code_str) <= 3:
        return code_str
    elif code_str.startswith('97'):
        return code_str[:3]
    else:
        return code_str[:2]

# Apply the function to create the new 'dept' column
df_filtered['dept'] = df_filtered['lieuexecution_code'].apply(determine_dept)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np

# Group by 'dept' and sum the 'montant' values
montant_sum_by_dept = df_filtered.groupby('dept')['montant'].sum().reset_index()
montant_sum_by_dept.columns = ['dept', 'montant_sum']

# Print the columns of montant_sum_by_dept to ensure the column names are correct
print("Columns of montant_sum_by_dept:", montant_sum_by_dept.columns)

# Print the columns of gdf to ensure the column names are correct
print("Columns of gdf before merge:", gdf.columns)

# Merge the summed 'montant' values with the GeoDataFrame
gdf = gdf.merge(montant_sum_by_dept, left_on='code_insee', right_on='dept', how='left')

# Print the columns of gdf after merge to check if 'montant_sum' is present
print("Columns of gdf after merge:", gdf.columns)

# Check if the merge was successful and the 'montant_sum' column exists
print(gdf.head())

# Apply logarithmic transformation to 'montant_sum'
if 'montant_sum' in gdf.columns:
    gdf['log_montant_sum'] = np.log1p(gdf['montant_sum'])  # Use log1p to handle zero values
else:
    raise KeyError("The 'montant_sum' column is not found in the GeoDataFrame.")

# Plot the map and zoom in to mainland France
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.boundary.plot(ax=ax)
gdf.plot(column='log_montant_sum', ax=ax, legend=True, cmap='OrRd', missing_kwds={"color": "lightgrey"})
plt.title('Log scale sum of montant by dept on a map of France')

# Set the extent to zoom in to mainland France
ax.set_xlim(-5, 10)
ax.set_ylim(41, 52)

plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Group by 'dept' and sum the 'montant' values
montant_sum_by_dept = df_filtered.groupby('dept')['montant'].sum().reset_index()
montant_sum_by_dept.columns = ['dept', 'montant_sum']

# Print the columns of montant_sum_by_dept to ensure the column names are correct
print("Columns of montant_sum_by_dept:", montant_sum_by_dept.columns)

# Print the columns of gdf to ensure the column names are correct
print("Columns of gdf before merge:", gdf.columns)

# Merge the summed 'montant' values with the GeoDataFrame
gdf = gdf.merge(montant_sum_by_dept, left_on='code_insee', right_on='dept', how='left')

# Print the columns of gdf after merge to check if 'montant_sum' is present
print("Columns of gdf after merge:", gdf.columns)

# Check if the merge was successful and the 'montant_sum' column exists
print(gdf.head())

# Calculate the 90th percentile of 'montant_sum'
if 'montant_sum' in gdf.columns:
    vmax = gdf['montant_sum'].quantile(0.90)
else:
    raise KeyError("The 'montant_sum' column is not found in the GeoDataFrame.")

# Plot the map and zoom in to mainland France
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.boundary.plot(ax=ax)
gdf.plot(column='montant_sum', ax=ax, legend=True, cmap='OrRd', vmax=vmax, missing_kwds={"color": "lightgrey"})
plt.title('Sum of montant with cut off, by dept on a map of France')

# Set the extent to zoom in to mainland France
ax.set_xlim(-5, 10)
ax.set_ylim(41, 52)

plt.show()
```
