This notebook just opens the main marchés public data, filters for Marseille and then merges with the pcv file. It can be adapted to any data and / or region. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import textwrap
```

```python
df = pd.read_csv('marches_publics_normalized_data.csv', delimiter=';', error_bad_lines=False)
```

```python
marseille_df = df[df['nom'].str.contains('Bouches|Marseille', case=False, na=False)]
marseille_df.head()   
```

```python
cpv_2008_df = pd.read_excel('cpv_2008_fr.xls')
cpv_2008_df.head()

merged_df = pd.merge(marseille_df, cpv_2008_df, left_on='codecpv', right_on='CODE', how='left')

merged_df.head()
```
